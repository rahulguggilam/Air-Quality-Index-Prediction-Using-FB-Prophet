Two Taks can be performed on the data set
1. Forecasting
2. Regression

```python
date=pd.to_datetime(air_quality['Date'],format='mixed')
date
```

```python

time_info=air_quality['Time'].apply(lambda x:x.replace('.',':'))
time_info
```

```python
ds=pd.concat([date,time_info],axis=1)
ds
```

```python
ds['ds']=ds['Date'].astype(str)+" "+ds['Time'].astype(str)


ds
```

```python
ds.drop(columns=["Date","Time"],axis=1,inplace=True)
```

```python
ds['ds']=pd.to_datetime(ds['ds'])
ds
```

```python
ds['y']=air_quality['RH']
ds
```

```python
from prophet import Prophet
```

```python
model=Prophet()
model.fit(ds)
```

```python
future=model.make_future_dataframe(periods=365,freq='H')
future.tail()
```

```python
forecast=model.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
model.plot(forecast)
```

```python
model.plot_components(forecast)
```
